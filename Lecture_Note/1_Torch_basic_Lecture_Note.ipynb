## Numpy <-> Pytorch 로 전환 자유롭게 하는 방법에 대해서

```python
import numpy as np
import torch

a = np.array([1,2,3])
b = torch.tensor([1,2,3])

a = torch.tensor(a)
print(a.type)
print(a.requires_grad)
```

```python
x = torch.tensor([1.0], requires_grad=True)
y = x**2 # x^2 으로 선언한 상황
print(y)

print(x.grad)
y.backward() # backward() 함수로 미분이 가능하다 - 미분하면 2x가 된 것을 확인하려고 한다
print("---after backward()---")
print(x.grad)
```

```python
x = torch.tensor([1.], requires_grad=True)
y = x**2
print(y)
y.retain_grad() # retain_grad()를 사용하는 위치가 중요하다 / 기본적으로 requires_grad=True인 애들만 미분 결과를 저장한다

z = 3*y
print(z)

z.backward()
print(x.grad)

# Warning이 발생하는 y.grad - retain_grad()를 적용해야 미분 값을 확인할 수 있다
# dz / dy 의 결과를 가지고 있다고 생각하면 되겠다
# grad_fn = 마지막에 한 행위를 기준으로 생각하기
print(y.grad)
```

## 간단한 인공신경망 만들기

```python
import torch
from torch import nn

x=torch.tensor([1.])
model = nn.Linear(1,1) # input layer - 1, outputl layer -1 개를 의미한다
print('-----')
print(model.weight)
print(model.bias)
print('-----')
y = model(x)
print(y)
print('-----')

# 기본 동작 원리는 다음처럼 표현이 가능하다
y2 = x @ model.weight + model.bias
print(y2)
print(y==y2)
```

```python
# fully-connected 한 모델에 대해서
fc1 = nn.Linear(1,3)
fc2 = nn.Linear(3,1)

# print(fc1.weight)
# print(fc1.bias)

# print(fc2.weight)
# print(fc2.bias)

x=torch.tensor([1.])

y1 = fc1(x) # 1 -> 3개의 output 출력
print(y1)
y2 = fc2(y1) # 3 -> 1개의 output 출력
print(y2)

y22 = (x@fc1.weight.T + fc1.bias)@fc2.weight.T + fc2.bias
print(y22)
print(y2 == y22)
```

```python
class MyModel(nn.Module):
    # 메서드 정의하기
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(2,5) # 주머니에 넣겠습니다 -> self 활용하기
        self.fc2 = nn.Linear(5,10)
        self.fc3 = nn.Linear(10,3)

        # sigmoid Activation
        self.sig1 = nn.Sigmoid()
        self.sig2 = nn.Sigmoid()
        self.sig3 = nn.Sigmoid()

    def forward(self, x):
        # 2 x 5 형태를 살펴보는 상황 - weight, bias 반영한다 - 5개 짜리에 sigmoid 함수가 활성화 함수 형태로 들어가 있다고 생각하면 된다
        x = self.fc1(x)
        x = self.sig1(x)
        # 모두 통과하려고 한다면 fc2 - fc3 로 연속으로 이어져야 한다
        x = self.fc2(x)
        x = self.sig2(x)
        x = self.fc3(x)
        x = self.sig3(x)
        return x

# instance 만들기
model = MyModel()

x= torch.randn(5,2)
y = model(x)
print(y)
```

```python
class SequentialModel(nn.Module):
    # 메서드 정의하기
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(2,5)
        self.fc2 = nn.Linear(5,10)
        self.fc3 = nn.Linear(10,3)

        self.sig1 = nn.Sigmoid()
        self.sig2 = nn.Sigmoid()
        self.sig3 = nn.Sigmoid()

        # 한번에 이어지도록 풀칠하는 과정 by nn.Sequential() 활용하기
        self.SeqModel = nn.Sequential(self.fc1,
                                      self.sig1,
                                      self.fc2,
                                      self.sig2,
                                      self.fc3,
                                      self.sig3)

    def forward(self, x):
        x = self.SeqModel(x)
        return x

x = torch.randn(5,2)
SeqModel = SequentialModel()
print(SeqModel.SeqModel[0])

# Activation 함수는 weight, bias를 주로 가지지 않지만
# 특정한 경우에는 가지고 있다 - ReLu 함수의 경우 parametric ReLu
# print(SeqModel.SeqModel[-1].bias) - error

# print(list(SeqModel.parameters()))

```

```python
# weight 초기화에 대해서 내가 직접 설정이 가능하다
# nn.init~ 을 활용하면 된다
# 내 필기용 : He initialization 이 공식 문서랑 다르다? => paper에 맞게 구현됐고 torch 공식 문서만 틀림
import torch
from torch import nn

Fin=5000
Fout=1000
w = torch.zeros(141, Fin)
nn.init.kaiming_uniform_(w, mode='fan_in', nonlinearity='relu') # forward pass에서 값의 범위를 유지시켜주기 위함
print(w.std())
print(torch.sqrt(torch.tensor(2/Fin)))
w = torch.zeros(Fout, 212)
nn.init.kaiming_uniform_(w, mode='fan_out', nonlinearity='relu') # backward pass에서 값의 범위를 유지시켜주기 위함
print(w.std())
print(torch.sqrt(torch.tensor(2/Fout)))

# CNN?
N=32
C=64
H=6
W=10
w = torch.zeros(N,C,H,W)
nn.init.kaiming_uniform_(w, mode='fan_in', nonlinearity='relu')
print(w.std())
print(torch.sqrt(torch.tensor(2/(C*H*W))))

w = torch.zeros(N,C,H,W)
nn.init.kaiming_uniform_(w, mode='fan_out', nonlinearity='relu')
print(w.std())
print(torch.sqrt(torch.tensor(2/(N*H*W))))
```
